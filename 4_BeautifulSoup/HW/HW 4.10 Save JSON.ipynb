```python
import requests
from bs4 import BeautifulSoup
import csv

def cook_soup(url):
    ans = requests.get(url)
    ans.encoding = 'utf-8'
    if ans.status_code != 200:
        print(ans.status_code, '\n\n', ans.text)
    return BeautifulSoup(ans.text, 'html.parser')
```

### Сбор данных о HDD в JSON
step 4

```python
import json
```

```python
soup = cook_soup('https://parsinger.ru/html/index4_page_1.html')
```

```python
"""
{
    "Наименование": "Toshiba S300 Surveillance",
    "Бренд": "Toshiba",
    "Форм-фактор": "3.5",
    "Ёмкость": "4 Tb",
    "Объем буферной памяти": "128 Mb",
    "Цена": "10710 руб"
}
"""
```

```python
list_j = []
for n in range(1, 5):
    url = f'https://parsinger.ru/html/index4_page_{n}.html'
    soup = cook_soup(url)
    for page in soup.find_all('div', {'class': 'img_box'}):
        card = {}
        card['Наименование'] = page.find('a', {'class': 'name_item'}).text.strip()
        for li in page.find_all('li'):
            card[li.text.split(':')[0]] = li.text.split(':')[1].strip()
        card['Цена'] = page.find('p', class_='price').text.strip()
        list_j.append(card)
```

```python
with open('Parse_data_about_HDD_for_json.json', 'w', encoding='utf-8') as f:
    json.dump(list_j, f, indent=4, ensure_ascii=False)
```

### Сбор данных со всех карточек товара

```python
list_j = []
for i in range(1, 6):
    for p in range(1, 5):
        url = f'https://parsinger.ru/html/index{i}_page_{p}.html'
        soup = cook_soup(url)
        for page in soup.find_all('div', {'class': 'img_box'}):
            card = {}
            card['Наименование'] = page.find('a', {'class': 'name_item'}).text.strip()
            for li in page.find_all('li'):
                card[li.text.split(':')[0]] = li.text.split(':')[1].strip()
            card['Цена'] = page.find('p', class_='price').text.strip()
            list_j.append(card)
```

```python
with open('Parse_all_data_for_json.json', 'w', encoding='utf-8') as f:
    json.dump(list_j, f, indent=4, ensure_ascii=False)
```

### Сбор данных о телефонах с карточек товара
Step 6

```python
ans = []
list_urls = []
main_url = 'https://parsinger.ru/html/'
for n in range(1, 5):
    url = f'https://parsinger.ru/html/index2_page_{n}.html'
    list_urls.append(url)
    
for url in list_urls:
    soup = cook_soup(url)
    for el in soup.find_all('div', {'class': 'item'}):
        card = {}
        href_val = el.find('a', {'class': 'name_item', 'href': True}).get('href')
        page_soup = cook_soup(main_url+href_val)
        card['categories'] = 'mobile'
        card['name'] = page_soup.find('p', {'id': 'p_header'}).text.strip()
        card['article'] = page_soup.find('p', {'class': 'article'}).text.split(':')[1].strip()
        dic_li = {}
        for li in page_soup.find_all('li'):
            dic_li[li.get('id')] = li.text.split(':')[1].strip()
        card['description'] = dic_li 
        card['count'] = page_soup.find('span', {'id': 'in_stock'}).text.split(':')[1].strip()
        card['price'] = page_soup.find('span', {'id': 'price'}).text.strip()
        card['old_price'] = page_soup.find('span', {'id': 'old_price'}).text.strip()
        card['link'] = main_url+href_val
        ans.append(card)        
```

```python
with open('Parse_phone_card_for_json.json', 'w', encoding='utf-8') as f:
    json.dump(ans, f, indent=4, ensure_ascii=False)
```
